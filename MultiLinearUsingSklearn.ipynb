```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set
from sklearn.linear_model import LinearRegression
```

```python
file_path = r'C:\Users\ajay4\OneDrive\Documents\Resume\Data science course\1.02.+Multiple+linear+regression.csv'
data = pd.read_csv(file_path)
data
```

```python
data.describe()
```

```python
x = data[['SAT','Rand 1,2,3']]
y = data['GPA']
```

## Linear Regression

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

## Calculating R squared (R^2)

```python
reg.score(x,y)
```

## Calculating Adjusted by its formula

$R^2_{adj.} = 1 - (1-R^2)*\frac{n-1}{n-1-p}$

```python
r2 = reg.score(x,y)
n = x.shape[0]
p = x.shape[1]
adjusted_r2 = 1-(1-r2)*(n-p)/(n-1-p)
adjusted_r2
```

# Feature Selection

```python
from sklearn.feature_selection import f_regression
```

```python
f_regression(x,y)
```

```python
p_values = f_regression(x,y)[1]
p_values
```

```python
p_values.round(3)
```

## Creating summary table

```python
reg_summary = pd.DataFrame(data = x.columns.values, columns = ['Feature'])
reg_summary
```

```python
reg_summary['coefficients'] = reg.coef_
reg_summary['pvalues'] = p_values.round(3)
reg_summary
```

## Standartization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(x)
scaler
x_scaler = scaler.transform(x)
x_scaler
```

## Regression with Scaler feature

```python
reg = LinearRegression()
reg.fit(x_scaler,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

## Creating summary table

```python
reg_summary = pd.DataFrame([['Intercept'],['SAT'],['Rand 1,2,3']], columns=['Feature'])
reg_summary['Weight'] = reg.intercept_ , reg.coef_[0], reg.coef_[1]
reg_summary
```

## Making predictions with the standartized coeffients

```python
new_data = pd.DataFrame(data=[[1700,2],[1800,1]],columns=['SAT','Rand 1,2,3'])
new_data 
reg.predict(new_data)
```

```python
new_data_scaled = scaler.transform(new_data)
new_data_scaled
```

```python
reg.predict(new_data_scaled)
```

## What if Rand 1,2,3 removed

```python
reg_simple = LinearRegression()
x_simple_matrix = x_scaler[:,0].reshape(-1,1)
reg_simple.fit(x_simple_matrix,y)
reg_simple.predict(new_data_scaled[:,0].reshape(-1,1))
```
